```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fdi_data = pd.read_csv("fdi_data.csv")
```

```python
print(fdi_data.isnull().sum())
```

```python
print(fdi_data.head())
```

```python
sector_wise_investment = fdi_data.groupby('Sector')['2000-01'].sum().reset_index()
print(sector_wise_investment)
```

```python
year_wise_investment = fdi_data.sum(axis=0).reset_index()
year_wise_investment.columns = ['Financial_Year', 'MINING']
print(year_wise_investment)
```

```python
year_wise_investment = fdi_data.sum(axis=0).reset_index()
year_wise_investment.columns = ['Financial_Year', 'Total_Investment']
print(year_wise_investment)
```

```python

financial_years = list(fdi_data.columns[1:])
melted_data = fdi_data.melt(id_vars=['Sector'], value_vars=financial_years, var_name='Financial_Year', value_name='Total_Investment')
plt.figure(figsize=(10, 6))
sns.lineplot(data=melted_data, x='Financial_Year', y='Total_Investment', hue='Sector')
plt.title('Year-wise FDI Investment in India by Sector')
plt.xlabel('Financial Year')
plt.ylabel('FDI Investment (in million USD)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


```
